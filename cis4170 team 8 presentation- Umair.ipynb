```python
import pandas as pd

# Load the dataset
file_path = 'Jobs_NYC_Postings_20240407.csv'
jobs_data = pd.read_csv(file_path)

# Display the first few rows of the dataset and the summary of the data
jobs_data.head(), jobs_data.info(), jobs_data.describe(include='all')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pie Chart for Geographical Distribution of Jobs
# Filter top 10 locations by job postings for clearer visibility in a pie chart
top_locations_pie = jobs_data['Work Location'].value_counts().head(10)

# Create pie chart
plt.figure(figsize=(10, 8))
plt.pie(top_locations_pie, labels=top_locations_pie.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Pie Chart of Job Postings by Top 10 Locations')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Horizontal Bar Chart for the same data
plt.figure(figsize=(12, 8))
sns.barplot(x=top_locations_pie.values, y=top_locations_pie.index, palette='coolwarm')
plt.title('Horizontal Bar Chart of Job Postings by Top 10 Locations')
plt.xlabel('Number of Job Postings')
plt.ylabel('Work Locations')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
data = {
    'Agency': ['Agency 1', 'Agency 2', 'Agency 3'],
    'Division A': [10, 20, 30],
    'Division B': [20, 15, 25],
    'Division C': [30, 25, 5]
}
df = pd.DataFrame(data)

# Setting the index to 'Agency'
df.set_index('Agency', inplace=True)

# Plotting a stacked bar chart
df.plot(kind='bar', stacked=True, figsize=(10, 7))
plt.title('Stacked Bar Chart of Divisions in Top Agencies')
plt.xlabel('Agency')
plt.ylabel('Number of Job Postings')
plt.legend(title='Division/Work Unit')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'jobs_data' is your DataFrame and it's already loaded
# Selecting a specific agency
specific_agency = 'DEPARTMENT OF TRANSPORTATION'
agency_data = jobs_data[jobs_data['Agency'] == specific_agency]

# Count job postings by division within the selected agency
division_counts = agency_data['Division/Work Unit'].value_counts().head(20)  # Top 20 divisions for better visibility

# Bar chart for the internal structure of the specific agency
plt.figure(figsize=(14, 8))
sns.barplot(x=division_counts.values, y=division_counts.index, palette='Spectral')
plt.title(f'Bar Chart of Divisions in {specific_agency} (Top 20 Divisions)')
plt.xlabel('Number of Job Postings')
plt.ylabel('Division/Work Unit')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
